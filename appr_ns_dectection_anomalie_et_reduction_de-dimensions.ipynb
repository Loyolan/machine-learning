```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
images = digits.images
X = digits.data
y = digits.target

from sklearn.ensemble import IsolationForest
model = IsolationForest(random_state=0, contamination = 0.02)
model.fit(X)
outliers = model.predict(X) == -1
plt.imshow(images[outliers][10])
plt.title(y[outliers][10])
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2)
X_reduced = model.fit_transform(X)
#X_reduced.shape
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
plt.colorbar()
```

```python
import numpy as np
model = PCA(n_components=64)
X_reduced = model.fit_transform(X)
plt.plot(np.cumsum(model.explained_variance_ratio_))
```

```python
np.argmax(np.cumsum(model.explained_variance_ratio_)>0.99)
```

```python
model = PCA(n_components=40)
X_reduced = model.fit_transform(X)

X_recovered = model.inverse_transform(X_reduced)
plt.subplot(121)
plt.imshow(X_recovered[100].reshape((8,8)))
plt.subplot(122)
plt.imshow(X[100].reshape((8,8)))
```

```python
model = PCA(n_components=0.99)
X_reduced = model.fit_transform(X)

X_recovered = model.inverse_transform(X_reduced)
plt.subplot(121)
plt.imshow(X_recovered[100].reshape((8,8)))
plt.subplot(122)
plt.imshow(X[100].reshape((8,8)))
```
