```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

plt.scatter(X[:,0], X[:,1], X[:,2], c=y)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(X)
model.labels_
```

```python
plt.figure(figsize=(11,6))
#lt.subplot(121)
plt.scatter(X[:,0], X[:,1], X[:,2], c=model.predict(X))
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], model.cluster_centers_[:,2], c="r")
#lt.subplot(122)
#lt.scatter(X[:,0], X[:,1], X[:,2], c=y)
```

```python
inertia = []
K_range = range(1, 20)
for k in K_range:
    model = KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)
plt.plot(K_range, inertia)
plt.xlabel("NB clusters")
plt.ylabel("Cout du model (Inertia)")
```

```python
#DETECTION D'ANOMALIE
from sklearn.ensemble import IsolationForest
model = IsolationForest(contamination=0.01)
model.fit(X)
plt.scatter(X[:,0], X[:, 1], X[:, 2], c=model.predict(X))
```
